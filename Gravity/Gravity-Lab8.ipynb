First enter the parameters you read from your graph for different cells and then fit the data, and plot the resulting fit on the same graph.  The cell below does a linear fit, and the following cell does a log plot and a power-law fit.  Simply put in your values for x and y and then the code will do the rest.
```python
y=[1,2,3]
x=[4,5,6]

p=np.polyfit(x,y,1)

yfit=p[0]*np.asarray(x)+p[1]

plt.figure()
plt.plot(x,y,'o',x,yfit)

plt.show()
print('yfit='+str('%1.2f'%p[0])+'x' + str('%1.2f' % p[1]))
```

```python
y=[1,4,9,16]
x=[1,2,3,4]

x_proc=np.log10(x)
y_proc=np.log10(y)

p=np.polyfit(x_proc,y_proc,1)

yfit=10**p[1]*x**p[0]

plt.figure()
plt.loglog(x,y,'o',x,yfit)

plt.show()
print('yfit=10^'+str('%1.2f'%p[1])+'x^'+str('%1.2f'%p[0]) )
```
